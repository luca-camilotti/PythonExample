```python
def divisori(m):
    return [d for d in range(2, m, 1) if m%d==0]

def divisibile(e, divlist):
    return [e//i for i in divlist if e%i==0]

def public(p, q, limit):
    divlist = divisori(p-1)+divisori(q-1)
    return [e for e in range(2, limit, 1) if len(divisibile(e, divlist))==0]

def private(p, q, e, limit):    
    return [(x+1)//e for x in range((p-1)*(q-1), limit*e, (p-1)*(q-1)) if (x+1)%e==0 ]

def crypt(value, n, key):
    return (value**key)%n
```
